# 5. Envío promedio por tienda

```python
# Análisis por categoría y producto
for i, tienda in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i}: Ventas por categoria")
    print(tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False))
    print('\n')

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i}: Productos más vendidos (cantidad)")
    print(df['Producto'].value_counts().head(10))
    print(f"Tienda {i}: Productos con mayor facturación")
    print(df.groupby('Producto')['Precio'].sum().sort_values(ascending=False).head(10))
    print('\n')
```

```python
# Métodos de pago
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i}: Distribución métodos de pago")
    print(df['Método de pago'].value_counts(normalize=True) * 100)
    print('\n')
```

```python

# Visualización geográfica
import matplotlib.pyplot as plt

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    plt.scatter(df['lon'], df['lat'], alpha=0.3, label=f'Tienda {i}')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Distribución geográfica de ventas')
plt.legend()
plt.show()
```

```python
# Evolución temporal
import pandas as pd
import matplotlib.pyplot as plt

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], format='%d/%m/%Y')
    ventas_mensuales = df.groupby(df['Fecha de Compra'].dt.to_period('M'))['Precio'].sum()
    ventas_mensuales.plot(kind='line', label=f'Tienda {i}', figsize=(12,6))
plt.title('Evolución mensual de ingresos por tienda')
plt.ylabel('Ventas mensuales')
plt.xlabel('Mes')
plt.legend()
plt.show()
```

```python
# Análisis de satisfacción
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i}: Calificación promedio general")
    print(df['Calificación'].mean())
    print(f"Tienda {i}: Calificación promedio por categoría")
    print(df.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False))
    print('\n')
```

```python
# Valores atípicos
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i}: Valores atípicos en precios")
    q1 = df['Precio'].quantile(0.25)
    q3 = df['Precio'].quantile(0.75)
    iqr = q3 - q1
    outliers = df[(df['Precio'] < q1 - 1.5 * iqr) | (df['Precio'] > q3 + 1.5 * iqr)]
    print(outliers[['Producto', 'Precio']].head())
    print('\n')
```
