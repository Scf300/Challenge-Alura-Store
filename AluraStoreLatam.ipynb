###**Comparación fracturación vs ventas por tienda**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Combine the total sales quantity by category for each store
df_cantidad = pd.DataFrame({
    'Tienda 1': total_venta_por_categoria.values,
    'Tienda 2': total_venta_por_categoria_2.values,
    'Tienda 3': total_venta_por_categoria_3.values,
    'Tienda 4': total_venta_por_categoria_4.values
}, index=total_venta_por_categoria.index)

# Combine the total sales revenue by category for each store
df_facturacion = pd.DataFrame({
    'Tienda 1': total_facturacion_por_categoria.values,
    'Tienda 2': total_facturacion_por_categoria_2.values,
    'Tienda 3': total_facturacion_por_categoria_3.values,
    'Tienda 4': total_facturacion_por_categoria_4.values
}, index=total_facturacion_por_categoria.index)

# PASO CRÍTICO: Normalizar cada categoría a rango 0-100
# Para cada categoría, el máximo valor = 100, los demás proporcionalmente

# Normalizar cantidad
df_cantidad_norm = df_cantidad.div(df_cantidad.max(axis=1), axis=0) * 100

# Normalizar facturación
df_facturacion_norm = df_facturacion.div(df_facturacion.max(axis=1), axis=0) * 100

# Create figure
fig, ax = plt.subplots(figsize=(16, 8))

# Configurar posiciones
categorias = df_cantidad.index
num_categorias = len(categorias)

bar_width = 0.2
x_pos = np.arange(num_categorias)

colors_tiendas = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
tiendas = list(df_cantidad.columns)

# Dibujar barras SUPERPUESTAS
for i, (tienda, color) in enumerate(zip(tiendas, colors_tiendas)):
    offset = (i - 1.5) * bar_width

    # Barras de CANTIDAD (color sólido)
    ax.bar(x_pos + offset, df_cantidad_norm[tienda], bar_width * 0.95,
           label=f'{tienda} - Cantidad', color=color, alpha=0.8,
           edgecolor='black', linewidth=0.7)

    # Barras de FACTURACIÓN (patrón rayado, superpuesto)
    ax.bar(x_pos + offset, df_facturacion_norm[tienda], bar_width * 0.95,
           label=f'{tienda} - Facturación', color=color, alpha=0.3,
           edgecolor='black', linewidth=0.7, hatch='//')

# Configurar eje
ax.set_ylabel('Valor Normalizado (0-100)', fontsize=13, fontweight='bold')
ax.set_xlabel('Categoría del Producto', fontsize=13, fontweight='bold')
ax.set_title('Comparación: Cantidad vs Facturación (Escala 0-100)\nBarras sólidas=Cantidad | Barras rayadas=Facturación',
             fontsize=14, fontweight='bold', pad=20)
ax.set_ylim(0, 110)
ax.set_xticks(x_pos)
ax.set_xticklabels(categorias, rotation=45, ha='right', fontsize=11)

# Leyenda
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc='upper left', bbox_to_anchor=(1.01, 1), fontsize=9)

# Grid
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

# **6.Informe final**


##**Introducción**
El presente informe presenta un análisis exhaustivo de las cuatro tiendas propiedad del Sr. Juan con el objetivo de determinar cuál de ellas sería más conveniente vender. Para tomar esta decisión, se evaluaron múltiples factores clave del negocio: ingresos totales generados, desempeño por categorías de productos, satisfacción del cliente (calificaciones promedio), productos más y menos vendidos, y análisis comparativo de eficiencia de ventas. El análisis busca identificar la tienda con menor potencial estratégico para la operación conjunta del portafolio, considerando no solo números absolutos sino también patrones de comportamiento de ventas.​

##**Análisis de Ingresos Totales**
Los ingresos totales representan uno de los indicadores más importantes para evaluar el desempeño de cada tienda:​

**Tienda 1:** $1,150,880,400 **(mayor facturación)**

**Tienda 2:** $1,116,343,500

**Tienda 3:** $1,098,019,600

**Tienda 4:** $1,038,375,700 **(menor facturación)**

La Tienda 1 lidera claramente en generación de ingresos, superando a la Tienda 4 por más de $112 millones. Esta diferencia del 10.8% es significativa y demuestra que la Tienda 1 tiene una capacidad superior para generar ventas y mantener un flujo de caja robusto.​

##**Análisis de Categorías de Productos**
El análisis por categorías reveló un patrón consistente en las cuatro tiendas, donde las categorías con mayor facturación son:​

**Top 3 categorías más vendidas (por ingresos):**

- **Electrónicos** - Primera posición en todas las tiendas

- **Electrodomésticos** - Segunda posición en todas las tiendas

- **Muebles** - Tercera posición en todas las tiendas

**Categorías con menor facturación:**

- Libros (última posición en todas las tiendas)

- Artículos para el hogar

- Deportes y diversión



Este patrón uniforme indica que las cuatro tiendas operan en el mismo mercado con preferencias similares de clientes, lo cual significa que la diferencia en desempeño no se debe a la mezcla de productos sino a otros factores operativos, de ubicación o estrategia comercial.​

##**Análisis de Calificaciones de Clientes**
La satisfacción del cliente es un indicador crítico de la calidad del servicio y la experiencia de compra:​

- **Tienda 3:** 4.048 **(calificación más alta)**

- **Tienda 2:** 4.037

- **Tienda 4:** 3.996

- **Tienda 1:** 3.977 **(calificación más baja)**

```python
# Valores atípicos
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i}: Valores atípicos en precios")
    q1 = df['Precio'].quantile(0.25)
    q3 = df['Precio'].quantile(0.75)
    iqr = q3 - q1
    outliers = df[(df['Precio'] < q1 - 1.5 * iqr) | (df['Precio'] > q3 + 1.5 * iqr)]
    print(outliers[['Producto', 'Precio']].head())
    print('\n')
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Unir dataframes de tiendas
df_total = pd.concat([tienda, tienda2, tienda3, tienda4])

# Contar cantidad de ventas por ciudad
ventas_por_ciudad = df_total.groupby('Lugar de Compra').size()

# Seleccionar las 5 ciudades con menor y mayor cantidad ventas
ciudades_menores = ventas_por_ciudad.nsmallest(5)
ciudades_mayores = ventas_por_ciudad.nlargest(5)

# Combinar en un solo DataFrame para graficar
ventas_combinadas = pd.concat([ciudades_menores, ciudades_mayores])

# Ordenar los valores para mostrar barras ordenadas
ventas_combinadas = ventas_combinadas.sort_values()

# Colores personalizados, primero rojo para menores, luego verde para mayores
colores = ['red']*5 + ['green']*5

# Crear gráfico horizontal
plt.figure(figsize=(12,6))
plt.barh(ventas_combinadas.index, ventas_combinadas.values, color=colores)
plt.title('Top 5 ciudades con menores y mayores ventas')
plt.xlabel('Cantidad de ventas')
plt.ylabel('Ciudad')
plt.show()
```
