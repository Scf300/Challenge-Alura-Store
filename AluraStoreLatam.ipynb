## Grafico comparacion ventas totales tiendas

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos de ventas totales
Comparacion_ventas_totales_tiendas = [total_ventas, total_ventas_2, total_ventas_3, total_ventas_4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear DataFrame con una fila y una columna por tienda
df_comparacion = pd.DataFrame([Comparacion_ventas_totales_tiendas], columns=nombres_tiendas)

# Colores por tienda
colors = ['blue', 'green', 'red', 'purple']

# Graficar
ax = df_comparacion.plot(kind='bar', figsize=(10, 6), color=colors)

# Eliminar etiquetas del eje X
ax.set_xticks([])

# Etiquetas y título
plt.title('Ventas Totales - Comparación entre Tiendas')
plt.ylabel('Ventas Totales')

# Leyenda con nombres de tiendas
plt.legend(title='Tiendas', labels=nombres_tiendas)

plt.tight_layout()
plt.show()
```

# 7.**Datos extra**

```python
# Evolución temporal
import pandas as pd
import matplotlib.pyplot as plt

for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], format='%d/%m/%Y')
    ventas_mensuales = df.groupby(df['Fecha de Compra'].dt.to_period('M'))['Precio'].sum()
    ventas_mensuales.plot(kind='line', label=f'Tienda {i}', figsize=(12,6))
plt.title('Evolución mensual de ingresos por tienda')
plt.ylabel('Ventas mensuales')
plt.xlabel('Mes')
plt.legend()
plt.show()
```

```python
# Valores atípicos
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    print(f"Tienda {i}: Valores atípicos en precios")
    q1 = df['Precio'].quantile(0.25)
    q3 = df['Precio'].quantile(0.75)
    iqr = q3 - q1
    outliers = df[(df['Precio'] < q1 - 1.5 * iqr) | (df['Precio'] > q3 + 1.5 * iqr)]
    print(outliers[['Producto', 'Precio']].head())
    print('\n')
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Unir dataframes de tiendas
df_total = pd.concat([tienda, tienda2, tienda3, tienda4])

# Contar cantidad de ventas por ciudad
ventas_por_ciudad = df_total.groupby('Lugar de Compra').size()

# Seleccionar las 5 ciudades con menor y mayor cantidad ventas
ciudades_menores = ventas_por_ciudad.nsmallest(5)
ciudades_mayores = ventas_por_ciudad.nlargest(5)

# Combinar en un solo DataFrame para graficar
ventas_combinadas = pd.concat([ciudades_menores, ciudades_mayores])

# Ordenar los valores para mostrar barras ordenadas
ventas_combinadas = ventas_combinadas.sort_values()

# Colores personalizados, primero rojo para menores, luego verde para mayores
colores = ['red']*5 + ['green']*5

# Crear gráfico horizontal
plt.figure(figsize=(12,6))
plt.barh(ventas_combinadas.index, ventas_combinadas.values, color=colores)
plt.title('Top 5 ciudades con menores y mayores ventas')
plt.xlabel('Cantidad de ventas')
plt.ylabel('Ciudad')
plt.show()
```
