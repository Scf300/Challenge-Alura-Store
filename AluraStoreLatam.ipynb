```python
total_ventas_2 = tienda2["Precio"].sum()
print(total_ventas_2)
```

## Tienda 3

### Ventas totales

```python
total_ventas_3 = tienda3["Precio"].sum()
print(total_ventas_3)
```

## Tienda 4

### Ventas totales

```python
total_ventas_4 = tienda4["Precio"].sum()
print(total_ventas_4)
```

## Tienda con mayores ventas



```python
tienda_mayor_venta = max(total_ventas, total_ventas_2, total_ventas_3, total_ventas_4)
print(tienda_mayor_venta)
```

```python
Ventas_totales_tiendas = [total_ventas, total_ventas_2, total_ventas_3, total_ventas_4]

# Usando sorted() para crear una nueva lista ordenada
calculo_ventas_tiendas_mayor_a_menor = sorted(Ventas_totales_tiendas, reverse=True)

print(calculo_ventas_tiendas_mayor_a_menor)
```

```python
""" Mayores ventas en orden descendente =
    Tienda 1
    Tienda 2
    Tienda 3
    Tienda 4
"""
```

## Grafico comparacion ventas totales tiendas

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos de ventas totales
Comparacion_ventas_totales_tiendas = [total_ventas, total_ventas_2, total_ventas_3, total_ventas_4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear DataFrame con una fila y una columna por tienda
df_comparacion = pd.DataFrame([Comparacion_ventas_totales_tiendas], columns=nombres_tiendas)

# Colores por tienda
colors = ['blue', 'green', 'red', 'purple']

# Graficar
ax = df_comparacion.plot(kind='bar', figsize=(10, 6), color=colors)

# Eliminar etiquetas del eje X
ax.set_xticks([])

# Etiquetas y título
plt.title('Ventas Totales - Comparación entre Tiendas')
plt.ylabel('Ventas Totales')

# Leyenda con nombres de tiendas
plt.legend(title='Tiendas', labels=nombres_tiendas)

plt.tight_layout()
plt.show()
```
